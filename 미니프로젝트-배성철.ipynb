### 2. 데이터 전처리

##### 2-1. 교통사고 데이터 셋 전처리

```python
trafficAccident_2020_seoul = pd.read_csv('trafficAccident_2020_seoul.csv')
trafficAccident_2020_seoul.head()
```

```python
trafficAccident_2020_seoul['발생년월일시']=trafficAccident_2020_seoul['발생년월일시'].apply(lambda x: pd.to_datetime(str(x), format='%Y-%m-%d %H'))
trafficAccident_2020_seoul.head()
```

```python
trafficAccident_2020_seoul['발생년월일시']=pd.to_datetime(trafficAccident_2020_seoul['발생년월일시']).dt.floor('H')
trafficAccident_2020_seoul.head()
```

```python
trafficAccident_2020_seoul.rename(columns={'발생년월일시':'년월일시'}, inplace=True)
trafficAccident_2020_seoul.head()
```

```python
trafficAccident_2020_seoul.set_index(trafficAccident_2020_seoul['년월일시'], inplace=True)
trafficAccident_2020_seoul.head()
```

```python
trafficAccident_2020_seoul.drop('년월일시', axis=1)
trafficAccident_2020_seoul.head()
```

```python
import numpy as np
trafficAccident_2020_seoul=trafficAccident_2020_seoul.resample('H').apply({'사망자수':np.sum}, axis=1).fillna(0)
trafficAccident_2020_seoul.head()
```

```python
weather_2020.head()
```

```python
trafficAccident_2020_seoul.head()
```

```python
preparation_data=pd.merge(
    weather_2020,
    trafficAccident_2020_seoul,
    how='outer',
    left_index=True,
    right_index=True
)
preparation_data.head()
```

```python
preparation_data=preparation_data.fillna(0)
preparation_data.head()
```

```python
preparation_data['사망자수']=preparation_data['사망자수'].astype(int)
preparation_data.head()
```

```python
preparation_data.info()
```

```python
preparation_data.to_csv('preparation_data.csv', index=False)
```
